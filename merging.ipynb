### `Q-10:` Create the complete dataset by uisng the below datasets:

This dataset should have these below column names in order:
1. Id
2. Species
3. SepalLengthCm
4. SepalWidthCm
5. PetalLengthCm
6. PetalWidthCm

Also, the dataset should be shuffled means the `Id` column should not be in increasing or decreasing order. So, make a dataset which has the shuffled Id column. You can use `DataFrame.sample()` method to shuffle.

````python
petal=pd.read_csv("petal_all - petal_all.csv")
sepal=pd.read_csv("sepal_all - sepal_all.csv")
setosa=pd.read_csv("setosa - setosa.csv")
versicolor=pd.read_csv("versicolor - versicolor.csv")
virginica=pd.read_csv("virginica - virginica.csv")
````

````python
virginica.head()
````

````python
sepal.head()
````

````python
v=pd.merge(sepal,petal,on='Id').drop(columns=["Unnamed: 0_x","Unnamed: 0_y"])
vn=pd.merge(v,virginica,on="Id")
ver=pd.merge(versicolor,sepal,on="Id").drop(columns=["Unnamed: 0_y","Unnamed: 0_x"])
sat=pd.merge(setosa,petal,on="Id").drop(columns=["Unnamed: 0_x","Unnamed: 0_y"])
iris=pd.concat([vn,ver,sat]).drop(columns="Unnamed: 0").sample(150)
iris
````

````python
#Q9 solutionn
slm=iris["SepalLengthCm"].mean()
plm=iris["PetalLengthCm"].mean()
iris["SepalLengthCm"].plot(kind="hist")
````

### `Q-11:` Find out the maximum and minimum sepal width and petal width of Setosa and Versicolor. To do this:
- First create a dataset with merging the required datasets
- After that, use `groupby` to create groups based on the "Species" column.
