### `Q-1:` Using `groupby` make groups using the `"Pclass"` column and find out the average age and total number of missing values in the `"Age"` column for every group.

````python
pcls=titanic.groupby("Pclass")
for group in list(pcls.groups.keys()):
  print(f"class {group} : avg is {pcls.get_group(group)['Age'].mean()}")
````

````python
for group in list(pcls.groups.keys()):
  print(f"total number of missing value in each group {group} is {pcls.get_group(group)['Age'].isna().sum()}")
````

### `Q-2:` Using `groupby` make groups using the `"Pclass"` column and fill every group's `"Embarked"` column's missing values with the mode value of that group. After that, print every group's `"Embarked"` column's value counts in ascending order.

````python
for group in list(pcls.groups.keys()):
  print("group is ",group,pcls.get_group(group)['Embarked'].isna().value_counts(ascending=True))
````

````python
for group in list(pcls.groups.keys()):
  pcls.get_group(group)['Embarked'].fillna(pcls.get_group(group)['Embarked'].mode(),inplace=True)
````

````python
for group in list(pcls.groups.keys()):
  print("group is ",group,pcls.get_group(group)['Embarked'].isnull().value_counts(ascending=True))
````

### `Q-5:` Find batsman in below category-
* Highest score while chasing
* Best Strike rate while chasing and have faced 100+ balls


> Chasing mean team batting in second inning

````python
ipl=pd.read_csv("ipl_deliveries.csv")
````

````python
ipl
````

````python
#Highest score while chasing
ipl[ipl["innings"]==2].groupby(["ID","batter"])["batsman_run"].sum().sort_values(ascending=False).head()
````

````python
# Best Strike rate while chasing and have faced 100+ balls
temp2=ipl[ipl["extra_type"]!="Wides"]
temp3=temp2.groupby("batter").agg({"ballnumber":"count","batsman_run":"sum"})
temp3=temp3[temp3["ballnumber"]>=100]
temp3["SR"]=(temp3["batsman_run"]/temp3["ballnumber"]*100)
temp3.sort_values(by="SR",ascending=False)
````
